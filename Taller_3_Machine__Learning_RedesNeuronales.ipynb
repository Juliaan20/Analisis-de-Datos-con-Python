#Desarrolle una Red Neuronal Artificial con **Keras** y **Tensorflow** para la compuerta XOR. Donde tenemos dos entradas binarias (**1 ó 0**) y la salida es 1 si una de las entradas es verdadera (**1**) y la otra falsa (**0**). 
#Es decir que de cuatro combinaciones posibles, sólo dos tienen salida 1 y las otras dos serán 0, como vemos aquí:
#• XOR(0,0) = 0
#• XOR(0,1) = 1
#• XOR(1,0) = 1
#• XOR(1,1) = 0

```
#Importaciòn de las librerias:
import numpy as np
import pandas as pd

# Librerias del Keras
from keras.models import Sequential
from keras.layers.core import Dense
```

```
# cargamos las 4 combinaciones de entrada
DatosEntrenamiento = np.array([[0,0],[0,1],[1,0],[1,1]])
DatosEntrenamiento
 
# cargamos los resultados que se obtienen, en el mismo orden
DatosSalida = np.array([[0],[1],[1],[0]])
DatosSalida

#Modelo de Red Neuronal:
model = Sequential()

# Capa entrada 
model.add(Dense(8, input_dim=2, activation='sigmoid'))

#capa de salida:
model.add(Dense(1, activation='sigmoid'))


#Compilaciòn de la Neurona:
model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])

#Entrenamos el modelo:
model.fit(DatosEntrenamiento, DatosSalida, epochs=1000)
 
#Evaluamos el modelo:
scores = model.evaluate(DatosEntrenamiento, DatosSalida)
 
#Mostramos en pantalla el puntaje:
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
print (model.predict(DatosEntrenamiento).round())
```
